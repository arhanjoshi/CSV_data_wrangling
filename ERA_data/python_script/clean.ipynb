```python
import pandas as pd
import glob
import os

# print(os.getcwd())

csv_dir = '../raw_csv/'
csv_files = glob.glob(os.path.join(csv_dir, '*.csv'))

print('CSV file found:')
for f in csv_files:
    print(f)
```

```python
for files in csv_files:
    df = pd.read_csv(files)
    count_entries = len(df)
    print(f"{os.path.basename(files)}: {count_entries} entries.")
```

```python
for files in csv_files:
    df = pd.read_csv(files)
    print(df.shape)
```

```python
df = pd.read_csv(csv_files[0])
df.head()
```

```python
print(df.columns.to_list())
```

```python
cols_to_drop = [
    'Account', 'sponsorNameNotAsLink()', 'primeSponsorNotAsLink()', 'leadFinancialUnitNotAsLink()', 'AMT GCO', 'Obligated and Anticipated',
    'Award End Date (anticipated)', 'Teams', 'Status',
    'Authorized total', 'Proposed Total'
]

df_cleaned = df.drop(columns=[col for col in cols_to_drop if col in df.columns])

df_cleaned.head()
```

```python
cleaned_dfs = []

for files in csv_files:
    temp_df = pd.read_csv(files)
    temp_df = temp_df.drop(columns=[cols for cols in cols_to_drop if cols in temp_df.columns])
    cleaned_dfs.append(temp_df)

print(f"Cleaned {len(cleaned_dfs)} files.")
```

```python
combined_df = pd.concat(cleaned_dfs, ignore_index = True)
print(combined_df.shape)
combined_df.head()
```

```python
output_path = '../cleaned_csv/combined_cleaned_data.csv'
combined_df.to_csv(output_path, index = False)

print(f"saved cleaned and combined data to {output_path}")
```
